```python
import pandas as pd
import anthropic
```

```python
df = pd.read_csv("data.csv")
```

```python
anthropic_api_key = "" #insert key
client = anthropic.Anthropic(api_key=anthropic_api_key)
```

```python
MODEL = "claude-3-5-sonnet-20240620"
MAX_TOKENS = 2000
TEMPERATURE = 0.0  # Adjust as needed
```

```python
def predict(data):
    
    prompt = data["Input Prompt"]
        
    response = client.messages.create(
                model=MODEL,
                max_tokens=MAX_TOKENS,
                temperature=TEMPERATURE,
                messages=[
                    {"role": "user", "content": prompt}
                ]
            )
    
    content = response.content[0].text
    print(content)
    
    return content
```

```python
df.iloc[0]["Claude"]
```

```python
results= []
```

```python
for i in range(len(df)):
    data = df.iloc[i]
    print(data)
    
    res = predict(data)
    results.append(res)
    if res == data["Correct Answer"]:
        results.append(1)
    else:
        results.append(0)
    
```

```python
df["Claude"]  = results
```

```python
import pandas as pd
import anthropic
```

```python
anthropic_api_key = "" #insert key
client = anthropic.Anthropic(api_key=anthropic_api_key)
```

```python
MODEL = "claude-3-5-sonnet-20240620"
MAX_TOKENS = 2000
TEMPERATURE = 0.0  # Adjust as needed
```

```python
def predict(data):
    prompt= "Consider the following context and answer the question. Give me just the answer without any other text in your response."
    prompt+= "context: " + data['context']
    prompt+= "question: "+ data['question']
        
    response = client.messages.create(
                model=MODEL,
                max_tokens=MAX_TOKENS,
                temperature=TEMPERATURE,
                messages=[
                    {"role": "user", "content": prompt}
                ]
            )
    
    content = response.content[0].text
    print(content)
    # break
    
    return content
```

```python
results = []
```

```python
for i in range(len(df)):
    data = df.iloc[i]
    print(data)
    
    res = predict(data)
    results.append(res)
    
```
